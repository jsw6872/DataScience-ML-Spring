```python
import pandas as pd
import time
```

```python
train_df = pd.read_csv('./data/amex/train_data.csv')
```

```python
%%time
train_labels_df = pd.read_csv('./data/amex/train_labels.csv')
```

```python
%%time
test_df = pd.read_csv('./data/amex/test_data.csv')
```

```python
%%time
train_df.to_feather('train_data')
train_labels_df.to_feather('train_labels')
test_df.to_feather('test_data')
```

```python
train_df.S_2 = pd.to_datetime(train_df.S_2)
```

```python
all_cols = [c for c in list(train_df.columns) if c not in ['customer_ID','S_2']]
```

```python
# description
cat_features = ["B_30","B_38","D_114","D_116","D_117","D_120","D_126","D_63","D_64","D_66","D_68"]
```

```python
num_features = [col for col in all_cols if col not in cat_features]
```

```python
%%time
train_group_df = train_df.groupby("customer_ID")[num_features].agg(['mean', 'std', 'min', 'max', 'last'])
```

```python
train_group_df.head(2)
```

```python
%%time

train_group_df.columns = ['_'.join(x) for x in train_group_df.columns]

test_cat_agg = train_df.groupby("customer_ID")[cat_features].agg(['count', 'last', 'nunique'])
test_cat_agg.columns = ['_'.join(x) for x in test_cat_agg.columns]
```

```python
test_cat_agg
```

```python
missing_df = pd.DataFrame(test_cat_agg.isnull().sum().sort_values(ascending=False)/len(test_cat_agg), columns=['percentage'])
missing_df
```
